# Charts for Reult Interpretation

## Figure 1: Genome Assembly Quality Metrics

```python
import matplotlib.pyplot as plt

# Data
isolates = ["SRR33816918", "SRR33816926"]

genome_size = [5291366, 5579775]   # bp
gc_content = [57.5, 57.1]          # %
n50 = [215893, 280628]              # bp
contigs = [105, 105]                # count

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Genome Assembly Quality Metrics", fontsize=14)

# Genome size
axes[0, 0].bar(isolates, genome_size)
axes[0, 0].set_title("Genome Size (bp)")
axes[0, 0].set_ylabel("Base pairs")

# GC content
axes[0, 1].bar(isolates, gc_content)
axes[0, 1].set_title("GC Content (%)")
axes[0, 1].set_ylabel("Percent GC")

# N50
axes[1, 0].bar(isolates, n50)
axes[1, 0].set_title("N50 (bp)")
axes[1, 0].set_ylabel("Base pairs")

# Contig count
axes[1, 1].bar(isolates, contigs)
axes[1, 1].set_title("Number of Contigs")
axes[1, 1].set_ylabel("Count")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("Figures/Genome_Assembly_Quality.png", dpi=300)
plt.show()
```

## Figure 1b: CheckM genome quality comparison

```python
import matplotlib.pyplot as plt

# Data
isolates = ["SRR33816918", "SRR33816926"]
completeness = [100.0, 100.0]
contamination = [0.09, 0.22]

# Create figure
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
fig.suptitle("CheckM Genome Quality Assessment", fontsize=14)

# Completeness plot
axes[0].bar(isolates, completeness)
axes[0].set_title("Completeness (%)")
axes[0].set_ylim(0, 105)
axes[0].set_ylabel("Percentage")

# Contamination plot
axes[1].bar(isolates, contamination)
axes[1].set_title("Contamination (%)")
axes[1].set_ylabel("Percentage")

plt.tight_layout()
plt.savefig("Figure_2_CheckM_quality.png", dpi=300)
plt.show()
```

## Figure 3a: Antimicrobial resistance gene distribution

```python
import matplotlib.pyplot as plt

# AMR class counts (derived from TSVs)
classes = ["Beta-lactam", "Fosfomycin", "Efflux (quinolone/phenicol)"]

srr18_counts = [1, 1, 3]
srr26_counts = [1, 1, 3]

x = range(len(classes))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x, srr18_counts, width, label="SRR33816918")
plt.bar([i + width for i in x], srr26_counts, width, label="SRR33816926")

plt.xticks([i + width / 2 for i in x], classes, rotation=15)
plt.ylabel("Number of AMR genes")
plt.title("Distribution of AMR Gene Classes")
plt.legend()

plt.tight_layout()
plt.savefig("Figure_3A_AMR_class_distribution.png", dpi=300)
plt.show()
```

## Figure 3b: Total AMR gene counts per isolate

## Figure 4: Non-AMR adaptive traits

```python
import matplotlib.pyplot as plt

categories = ["Metal resistance", "Heat stress", "Virulence"]

srr18 = [5, 0, 0]
srr26 = [18, 2, 2]

x = range(len(categories))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x, srr18, width, label="SRR33816918")
plt.bar([i + width for i in x], srr26, width, label="SRR33816926")

plt.xticks([i + width / 2 for i in x], categories, rotation=15)
plt.ylabel("Gene count")
plt.title("Stress, Metal, and Virulence-Associated Genes")
plt.legend()

plt.tight_layout()
plt.savefig("Figure_4_Adaptive_traits.png", dpi=300)
plt.show()
```
